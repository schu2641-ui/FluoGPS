```python
from rdkit import Chem
from rdkit.Chem.Draw import MolDraw2DSVG
from IPython.display import SVG

smiles = 'CCO'
molecule = Chem.MolFromSmiles(smiles)

# 创建 SVG 绘图器
d = MolDraw2DSVG(400, 400)

# 设置绘制选项 - 加粗键
d.drawOptions().bondLineWidth = 4

# 绘制分子
d.DrawMolecule(molecule)
d.FinishDrawing()

# 获取 SVG 并修改颜色
svg = d.GetDrawingText()
svg = svg.replace('stroke:#000000', 'stroke:#667F84')
svg = svg.replace('stroke:#0000FF', 'stroke:#667F84')
svg = svg.replace('stroke:#FF0000', 'stroke:#667F84')
SVG(svg)
```
